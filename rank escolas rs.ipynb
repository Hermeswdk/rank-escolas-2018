```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
#carregar microdados
df = pd.read_csv('../MICRODADOS_ENEM_2018.csv', encoding='Latin1', sep=';')
```

```python
#selecionar apenas quem participou ads quatro provas
df2 = df[(df['TP_PRESENCA_CN'] == 1) & (df['TP_PRESENCA_CH'] == 1) & (df['TP_PRESENCA_LC'] == 1) & (df['TP_PRESENCA_MT'] == 1)]
```

```python
#remover quem não esta cadastrado em nenhuma escola
df3 = df2.loc[df2['CO_ESCOLA'].notna()]
```

```python
#carregar e selecionar escolas apenas do Rio Grande do Sul
escolas = pd.read_csv('../ESCOLAS.CSV', encoding='Latin1', sep = '|')
escolas = escolas.loc[43 == escolas['CO_UF']]
escolas = escolas.loc[1 == escolas['IN_REGULAR']]
escolas = escolas[['CO_ENTIDADE','NO_ENTIDADE', 'CO_MUNICIPIO']]
```

```python
escolas['CO_ENTIDADE'].unique().tolist()
```

```python
#selecionar entre todos os participanes, os que estão cadastrados em alguma escola do RS
df4 = df3.loc[df3['CO_ESCOLA'].isin(escolas['CO_ENTIDADE'].unique().tolist())]
```

```python
df4.shape
```

```python
df4.head()
```

```python
#selecionar estudantes que marcaram no enem nos critérios "Estou cursando e concluirei o Ensino Médio em 2018." Q026 == 'B'
#e que estão matriculados no tipo de ensino regular
rs = df4.loc[df4['TP_ENSINO'] == 1]

rs2 = rs.loc[rs['Q026'] == 'B']
```

```python
rs2.shape
```

```python
#seguindo os critérios no INEP, seleciono apenas escolas que tiveram pelo menos 10 alunos no enem
#e salvo em outra coluna o total de alunos que essa escola teve no enem
rs3 = rs2.loc[rs2.groupby('CO_ESCOLA')['CO_ESCOLA'].transform('count') >= 10]

rs3.loc[:, 'no enem'] = rs3.groupby('CO_ESCOLA')['CO_ESCOLA'].transform('count')
```

```python
rs3.head()
```

```python
#Leio os dados de matriculas da região sul, e seleciono apenas as do estado do Rio Grande do Sul

mat = pd.read_csv('MATRICULA_SUL.CSV', encoding='Latin1', sep='|')

mat = mat.loc[mat['CO_UF'] == 43]
```

```python
#seleciono entre os estudantes, apenas os matriculados no terceiro ano do ensino médio e técnico
#assim como os matriculados no tipo de ensino REGULAR

ensMedioCods = [27, 32]
mat2 = mat.loc[mat['TP_ETAPA_ENSINO'].isin(ensMedioCods)]

#Verificar resutado
mat2 = mat2.loc[mat2['IN_REGULAR'] == 1]
```

```python
mat2.head()
```

```python
#descubro qual é o tamanho das turmas do terceiro ano de todas as escolas

terceiroAno = mat2['CO_ENTIDADE'].value_counts().reset_index()
terceiroAno.rename({'CO_ENTIDADE': 'total_matriculas', 'index': 'CO_ESCOLA'},axis = 1, inplace = True)

terceiroAno.head()
```

```python
terceiroAno.shape
```

```python
#uno os dados dos estudantes que fizeram o enem com os dados da sua escola
allData = rs3.merge(terceiroAno, on ='CO_ESCOLA')
```

```python
allData[['CO_ESCOLA', 'no enem', 'total_matriculas']].drop_duplicates().head()
```

```python
#remover as escolas que não tiveram pelo menos 50% dos matriculados no terceiro ano presentes no enem.
mask = allData['no enem'] >= (allData['total_matriculas'] // 2)
allData2 = allData.loc[mask]
```

```python
#calculo a média de cada estudante na prova objetiva e na prova objetiva + redação

allData2.loc[:, 'nota_final'] = allData2.loc[:, ['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO']].mean(1)

allData2.loc[:, 'prova_objetiva'] = allData2.loc[: ,['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT']].mean(1)
```

```python
allData2.head()
```

```python
#calculo a média da nota de cada escola(redação, provas objetivas e provas objetivas + redação)
rankEscolas = allData2.groupby('CO_ESCOLA')['NU_NOTA_REDACAO','prova_objetiva','nota_final'].mean().reset_index()
```

```python
rankEscolas['rank_redacao'] = rankEscolas['NU_NOTA_REDACAO'].rank(ascending =False)
rankEscolas['rank_prova_objetiva'] = rankEscolas['prova_objetiva'].rank(ascending =False)
rankEscolas['rank_nota_final'] = rankEscolas['nota_final'].rank(ascending =False)
```

```python
rankEscolas.sort_values('nota_final', ascending=False).head(20)
```

```python
#uno novamente o rank com o DF escolas para pegar o nome da escola e o código do município
rankEscolas = rankEscolas.merge(escolas, left_on='CO_ESCOLA', right_on='CO_ENTIDADE')
```

```python
rankEscolas.head()
```

```python
#pegar o nome do múnicípio

codes = pd.read_csv('../../../dados/rs_codes.csv')

codes = codes.loc[: ,['municipio', 'id_ibge']]

rankEscolas2 = rankEscolas.merge(codes, left_on='CO_MUNICIPIO', right_on='id_ibge')

rankEscolas2.head()
```

```python
#excluo colunas duplicadas pelos merges
#reordeno a ordem das colunas
#renomeio a coluna da redação
#e reordeno os dados pela coluna 'rank_nota_final'

rankEscolas2.drop(['CO_ESCOLA', 'CO_MUNICIPIO'], axis = 1, inplace = True)
rankEscolas2 = rankEscolas2.loc[:, ['CO_ENTIDADE', 'id_ibge', 'NO_ENTIDADE', 'municipio', 'NU_NOTA_REDACAO','prova_objetiva','nota_final','rank_redacao','rank_prova_objetiva','rank_nota_final']]
rankEscolas2.rename({'NU_NOTA_REDACAO': 'nota_redacao'}, axis = 1, inplace=True)

rankEscolas2.sort_values('rank_nota_final', ascending=True, inplace=True)
```
